```python
import subprocess
import re

def run_sampler(args):
    command = args
    process = subprocess.run(command, capture_output=True, text=True)

    pattern = r'total time for sampling: (\d+\.\d)'
    match = re.search(pattern, process.stderr) # FIXME
    if match:
        res = match.group(1)
        return res
    else:
        raise ValueError

```

```python

args = ['python', 'sampler.py', '-i', '../models/color.wfomcs', '-k', '100']
res = []
for _ in range(1):
    res.append(float(run_sampler(args)))
print(res)
print(sum(res)/len(res))
        

args = ['python', 'samplerv2.py', '-i', '../models/color.wfomcs', '-k', '100']
res = []
for _ in range(1):
    res.append(float(run_sampler(args)))
print(res)
print(sum(res)/len(res))
```
